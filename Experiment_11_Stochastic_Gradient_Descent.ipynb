# Experiment 11

## Aim

To implement Stochastic Gradient Descent (SGD) for Linear Regression and observe the convergence of the model.

## Theory

Stochastic Gradient Descent (SGD) is an optimization algorithm that updates the model parameters after processing **one training sample at a time**, unlike Batch Gradient Descent, which uses the entire dataset.

Weight Update Equation:

w = w − α × dw

Bias Update Equation:

b = b − α × db

where

α = Learning Rate

Advantages of SGD:
- Faster updates
- Efficient for large datasets
- Lower memory usage

Disadvantages:
- Noisy convergence
- Sensitive to learning rate

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Training Data

X = np.array([1,2,3,4,5], dtype=float)
Y = np.array([2,4,6,8,10], dtype=float)

plt.figure(figsize=(6,4))
plt.scatter(X, Y, color="red")
plt.title("Training Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
w = 0
b = 0

learning_rate = 0.01

epochs = 100

loss_history = []
```

```python
for epoch in range(epochs):

    total_loss = 0

    for i in range(len(X)):

        x = X[i]
        y = Y[i]

        y_pred = w*x + b

        error = y_pred - y

        dw = 2*error*x
        db = 2*error

        w = w - learning_rate*dw
        b = b - learning_rate*db

        total_loss += error**2

    loss_history.append(total_loss/len(X))

print("Final Weight =", round(w,4))
print("Final Bias =", round(b,4))
```

```python
plt.figure(figsize=(7,5))

plt.plot(loss_history,
         linewidth=2)

plt.title("SGD Loss Curve")

plt.xlabel("Epochs")

plt.ylabel("Loss")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.scatter(X, Y,
            color="red",
            label="Training Data")

plt.plot(X,
         w*X+b,
         color="blue",
         linewidth=2,
         label="Regression Line")

plt.title("Linear Regression using SGD")

plt.xlabel("X")

plt.ylabel("Y")

plt.legend()

plt.grid(True)

plt.show()
```

```python
new_x = np.array([6,7,8])

predictions = w*new_x + b

print("Predictions")

for x, pred in zip(new_x, predictions):
    print(f"Input = {x}, Predicted = {pred:.2f}")
```
