# Experiment 10

## Aim

To implement the Gradient Descent Algorithm for Linear Regression and analyze the convergence of the loss function.

## Theory

Gradient Descent is an optimization algorithm used to minimize the loss function by updating the model parameters (weight and bias) iteratively.

The update equations are:

Weight Update:

w = w − α(dw)

Bias Update:

b = b − α(db)

where

α = Learning Rate

The objective is to minimize the Mean Squared Error (MSE) between the predicted and actual values.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Training Data

X = np.array([1,2,3,4,5],dtype=float)

Y = np.array([2,4,6,8,10],dtype=float)

plt.scatter(X,Y,color="red")

plt.title("Training Data")

plt.xlabel("X")

plt.ylabel("Y")

plt.grid(True)

plt.show()
```

```python
w = 0

b = 0

learning_rate = 0.01

epochs = 100

loss_history = []

weight_history = []
```

```python
for i in range(epochs):

    y_pred = w*X + b

    loss = np.mean((Y-y_pred)**2)

    loss_history.append(loss)

    weight_history.append(w)

    dw = (-2/len(X))*np.sum(X*(Y-y_pred))

    db = (-2/len(X))*np.sum(Y-y_pred)

    w = w-learning_rate*dw

    b = b-learning_rate*db

print("Final Weight =",round(w,4))

print("Final Bias =",round(b,4))
```

```python
plt.figure(figsize=(7,5))

plt.plot(loss_history,
         color="blue",
         linewidth=2)

plt.title("Loss vs Iterations")

plt.xlabel("Iterations")

plt.ylabel("Loss")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.scatter(X,Y,
            color="red",
            label="Training Data")

plt.plot(X,
         w*X+b,
         color="blue",
         linewidth=2,
         label="Regression Line")

plt.legend()

plt.grid(True)

plt.title("Linear Regression using Gradient Descent")

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.plot(weight_history,
         color="green",
         linewidth=2)

plt.title("Weight Updates")

plt.xlabel("Iterations")

plt.ylabel("Weight")

plt.grid(True)

plt.show()
```
