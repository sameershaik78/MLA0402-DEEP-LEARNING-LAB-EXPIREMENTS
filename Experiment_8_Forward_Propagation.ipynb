## Aim

To implement Forward Propagation in a simple Neural Network using the Sigmoid activation function.

## Theory

Forward Propagation is the process of passing input data through the neural network to generate predictions.

The steps are:

1. Compute the weighted sum (z = Wx + b)
2. Apply the activation function
3. Pass the output to the next layer
4. Continue until the output layer

The Sigmoid activation function is:

σ(z) = 1 / (1 + e⁻ᶻ)

Its output ranges from 0 to 1, making it suitable for binary classification problems.

```python
import numpy as np
```

```python
# Input values
X = np.array([1, 2])

# Hidden layer weights
W1 = np.array([[0.2, 0.4],
               [0.5, 0.3]])

# Hidden layer bias
b1 = np.array([0.1, 0.2])

# Output layer weights
W2 = np.array([[0.6],
               [0.7]])

# Output bias
b2 = np.array([0.3])
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
z1 = np.dot(X, W1) + b1

a1 = sigmoid(z1)

print("Hidden Layer Output")

print(a1)
```

```python
z2 = np.dot(a1, W2) + b2

output = sigmoid(z2)

print("Final Output")

print(output)
```

```python
print("Input")

print(X)

print()

print("Hidden Layer Weighted Sum")

print(z1)

print()

print("Hidden Layer Activation")

print(a1)

print()

print("Output Layer Weighted Sum")

print(z2)

print()

print("Network Output")

print(output)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.bar(["Neuron 1", "Neuron 2"], a1)

plt.title("Hidden Layer Activation")

plt.ylabel("Activation Value")

plt.grid(axis="y")

plt.show()
```

```python
plt.figure(figsize=(4,4))

plt.bar(["Output"], output)

plt.title("Final Network Output")

plt.ylabel("Probability")

plt.ylim(0,1)

plt.grid(axis="y")

plt.show()
```
