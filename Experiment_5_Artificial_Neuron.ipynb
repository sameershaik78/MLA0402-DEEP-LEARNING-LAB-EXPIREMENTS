## Aim

To implement a single Artificial Neuron using Sigmoid and ReLU activation functions and compare their outputs.

## Theory

An Artificial Neuron is the basic building block of a Neural Network. It receives input values, multiplies them by weights, adds a bias, and passes the result through an activation function.

The output of a neuron is calculated as:

z = w₁x₁ + w₂x₂ + w₃x₃ + ... + b

Two commonly used activation functions are:

1. Sigmoid
2. ReLU

Sigmoid converts values into the range (0,1).

ReLU converts all negative values to zero while keeping positive values unchanged.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Input values
x = np.array([1, 2, 3])

# Weights
w = np.array([0.5, 0.5, 0.5])

# Bias
b = 0.5

print("Input :", x)
print("Weights :", w)
print("Bias :", b)
```

```python
z = np.dot(x, w) + b

print("Weighted Sum =", z)
```

```python
sigmoid = 1 / (1 + np.exp(-z))

print("Sigmoid Output =", sigmoid)
```

```python
relu = max(0, z)

print("ReLU Output =", relu)
```

```python
print("----------------------")
print("Weighted Sum :", z)
print("Sigmoid :", sigmoid)
print("ReLU :", relu)
```

```python
values = [sigmoid, relu]

labels = ["Sigmoid", "ReLU"]

plt.figure(figsize=(6,4))

plt.bar(labels, values,
        color=["green","orange"])

plt.title("Activation Function Comparison")

plt.ylabel("Output")

plt.grid(axis="y")

plt.show()
```

```python
x_values = np.linspace(-10,10,100)

sigmoid_curve = 1/(1+np.exp(-x_values))

plt.figure(figsize=(6,4))

plt.plot(x_values,
         sigmoid_curve,
         color="blue")

plt.title("Sigmoid Activation Function")

plt.xlabel("Input")

plt.ylabel("Output")

plt.grid(True)

plt.show()
```

```python
relu_curve = np.maximum(0,x_values)

plt.figure(figsize=(6,4))

plt.plot(x_values,
         relu_curve,
         color="red")

plt.title("ReLU Activation Function")

plt.xlabel("Input")

plt.ylabel("Output")

plt.grid(True)

plt.show()
```
